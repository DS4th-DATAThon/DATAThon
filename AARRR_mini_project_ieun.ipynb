## 문제 6: Retention:코호트 분석(고객)

```python
import seaborn as sns
retention_df = retail_df[(retail_df['InvoiceDate'] > '2010-01-01')].copy()
retention_df = retention_df.merge(active_customer_df, on='Customer ID')
retention_df[(retention_df['Active/InActive'] == 'Active')]
```

```python
retention_df['current_month'] = pd.to_datetime(retention_df['InvoiceDate']).dt.year.map(str) + "-" + pd.to_datetime(retention_df['InvoiceDate']).dt.month.map(str) + "-" '01'
retention_df['current_month'] = pd.to_datetime(retention_df['current_month']).dt.date
retention_df["cohort_month"] = retention_df.groupby("Customer ID")["current_month"].transform("min")

retention_df
```

```python
dates = pd.Series(['2011-12-01', '2011-08-01'])
dates = pd.to_datetime(dates)
print(dates.dt.quarter)
```

```python
def quarter_diff(df, col1, col2):
    current_quarter = pd.to_datetime(df[col1]).dt.quarter
    cohort_quarter = pd.to_datetime(df[col2]).dt.quarter
    return current_quarter - cohort_quarter

retention_df['quarter_diff']=quarter_diff(retention_df, 'current_month', 'cohort_month')
```

```python
def years_diff(df, col1, col2):
    current_year = pd.to_datetime(df[col1]).dt.year
    cohort_year = pd.to_datetime(df[col2]).dt.year
    return current_year - cohort_year

retention_df['years_diff'] = years_diff(retention_df, 'current_month', 'cohort_month')
```

```python
def cohort_index(df, col1, col2):
    return df[col1] * 4 + df[col2]

retention_df['cohort_index'] = cohort_index(retention_df, 'years_diff', 'quarter_diff')
```

```python
retention_df['cohort_month'] = pd.to_datetime(retention_df['cohort_month'])
#retention_df['cohort_label'] = pd.to_datetime(retention_df['cohort_month'], format="Q%Y").dt.to_period('Q')
retention_df['cohort_label'] = pd.to_datetime(retention_df['cohort_month']).dt.to_period('Q')
```

```python
retention_df['cohort_label']
```

```python
cohort_counts = retention_df.groupby(['cohort_label', 'cohort_index'])['Customer ID'].nunique()
cohort_counts_df = cohort_counts.to_frame().rename(columns={'Customer ID':'users'}).sort_values(by=['cohort_label'])
cohort_counts_df['retention_rate'] = cohort_counts_df['users'] / cohort_counts_df.groupby(["cohort_label"])["users"].transform("max")
cohorts_pivot = cohort_counts_df.pivot_table(index='cohort_label', columns = 'cohort_index', values = 'retention_rate')
cohorts_pivot
```

```python
cohort_counts_df['users']
```

```python
#히트맵 코드
plt.figure(figsize=(12, 8))

heatmap = sns.heatmap(
    data=cohorts_pivot,
    annot=True,                # 셀 내부에 값 표시
    fmt='.2%',                 # 텍스트 형식 (소수점 2자리)
    cmap='Greens',             # 색상 팔레트
    cbar_kws={'label': 'Retention Rate, %'},  # 컬러바 제목
    linewidths=0.5,            # 셀 간격
    linecolor='gray',          # 셀 경계 색상          # Retention Rate의 범위 설정
)
```

## 문제 7: Retention: 코호트 분석(평균 구매수량)

```python
avg_cohort_counts = retention_df.groupby(['cohort_label', 'cohort_index'])['Quantity'].mean()
avg_cohort_counts_df = avg_cohort_counts.to_frame().sort_values(by=['cohort_label'])
avg_cohort_counts_df['retention_rate'] = avg_cohort_counts_df['Quantity'] 
avg_cohorts_pivot = avg_cohort_counts_df.pivot_table(index='cohort_label', columns = 'cohort_index', values = 'retention_rate')
avg_cohorts_pivot
```

```python
#heatmap 코드
plt.figure(figsize=(12, 8))

heatmap = sns.heatmap(
    data=avg_cohorts_pivot,
    annot=True,                # 셀 내부에 값 표시
    fmt='.1f',                 # 텍스트 형식 (소수점 1자리)
    cmap='Blues',              # 색상 팔레트
    cbar_kws={'label': 'Average Quantity'},  # 컬러바 제목
    linewidths=0.5,            # 셀 간격
    linecolor='gray',          # 셀 경계 색상
    vmin=0                     # 최소값 설정 (필요 시 조정 가능)
)
```

## 문제10

```python
import numpy as np
from scipy.stats import uniform
waiting_times = uniform.rvs(loc=5, scale=10, size=100, random_state=0 )
print(np.shape(waiting_times))

mean_waiting_time = np.mean(waiting_times)
std_waiting_time = np.std(waiting_times, ddof=1)

print(f"평균 대기시간: {mean_waiting_time:.2f}")
print(f"표준편차: {std_waiting_time:.2f}")
```

## 문제11

```python
from scipy.stats import bernoulli

# 성공 확률과 시행수 정의
p = 0.3
n = 10

# 베르누이 분포 샘플 생성
outcomes = bernoulli.rvs(p, size=n)
success_count = outcomes.sum()

print(f"각 시도 결과: {outcomes}")
print(f"성공 횟수: {success_count}")
```

## 문제12

```python
from scipy.stats import binom

# 이항 분포: 문제 수 , 정답확률, 학생 수
num_of_stds = 20
num_of_quizzes = 5
percent = 0.7

# 각 학생의 점수 시뮬레이션
scores = binom.rvs(num_of_quizzes, p = percent, size=num_of_stds, random_state=0)
average_score = np.mean(scores)

print(f"학생별 점수: {scores}")
print(f"평균 점수: {average_score:.2f}")
```

## 문제13

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# 정규 분포 파라미터: 평균, 표준편차, 데이터 샘플 수
mu = 50
sigma = 5
num_of_samples = 1000

# 샘플 생성
weights = norm.rvs(loc=mu, scale=sigma, size=num_of_samples, random_state=0)

# 45g 이상 55g 이하의 비율 계산
conds = (weights >= 45) & (weights <=55)

within_range = len(weights[conds]) / len(weights) * 100
print(f'45g 이상 55g 이하의 비율 계산: {within_range:.2f}%')
```
